# Linear Search

```python
def linear_search(array: Sequence[int], target: int) -> int:
    for i in range(len(array)):
        if array[i] == target:
            return i
    return -1

linear_search([5, 1, 3, 7, 3, 2, 4, 6], 2)
```

# Binary Search

```python
def binary_search(array: Sequence[int], target: int) -> int:
    low = 0
    high = len(array) - 1
    while low <= high:
        mid = low + (high - low) // 2
        if array[mid] == target:
            return mid
        elif array[mid] < target:
            low = mid + 1
        elif array[mid] > target:
            high = mid - 1
    return -1

array = [1, 2, 3, 5, 6, 8, 9, 10, 12]
for i in range(len(array)):
    print(f"Searching {array[i]}:\t position found = {binary_search(array, array[i])}")
print(f"Searching 7:\t position found = {binary_search(array, 7)}")
```

# Interpolation Search

```python
def interpolation_search(array: Sequence[int], target: int) -> int:
    low = 0
    high = len(array) - 1
    while low <= high and target >= array[low] and target <= array[high]:
        mid = low + ((high - low) * (target - array[low])) // (array[high] - array[low])
        if array[mid] == target:
            return mid
        elif array[mid] < target:
            low = mid + 1
        elif array[mid] > target:
            high = mid - 1
    return -1

array = [1, 2, 3, 5, 6, 8, 9, 10, 12]
for i in range(len(array)):
    print(f"Searching {array[i]}:\t position found = {interpolation_search(array, array[i])}")
print(f"Searching 7:\t position found = {interpolation_search(array, 7)}")
```

# Jump Search

```python
def jump_search(array: Sequence[int], target: int) -> int:
    jump_size = int(len(array)**(0.5))
    low = 0
    high = jump_size
    while low < high:
        ret = linear_search(array[low: high+1], target)
        if ret != -1:
            return low + ret
        low += jump_size
        high += jump_size
        if high > len(array)-1:
            high = len(array) - 1
    return -1

array = [1, 2, 3, 5, 6, 8, 9, 10, 12]
for i in range(len(array)):
    print(f"Searching {array[i]}:\t position found = {jump_search(array, array[i])}")
print(f"Searching 7:\t position found = {jump_search(array, 7)}")
```

# Ternary Search

```python
def terenary_search(array: Sequence[int], target: int) -> int:
    low = 0
    high = len(array) - 1
    while low <= high:
        mid1 = low + (high - low) // 3
        mid2 = high - (high - low) // 3
        if array[mid1] == target:
            return mid1
        elif array[mid2] == target:
            return mid2
        elif array[mid1] < target:
            if array[mid2] < target:
                low = mid2 + 1
            else:
                low = mid1 + 1
                high = mid2 - 1
        elif array[mid1] > target:
            if array[mid2] > target:
                high = mid1 - 1
            else:
                high = mid2 - 1
                low = mid1 + 1
    return -1

array = [1, 2, 3, 5, 6, 8, 9, 10, 12]
for i in range(len(array)):
    print(f"Searching {array[i]}:\t position found = {terenary_search(array, array[i])}")
print(f"Searching 7:\t position found = {terenary_search(array, 7)}")
```

# Exponential Search

```python
def binary_search(array: Sequence[int], low: int, high: int, target: int) -> int:
    while low <= high:
        mid = low + (high - low) // 2
        if array[mid] == target:
            return mid
        elif array[mid] < target:
            low = mid + 1
        elif array[mid] > target:
            high = mid - 1
    return -1


def exponential_search(array: Sequence[int], target: int) -> int:
    i = 1
    while array[i-1] < target:
        i = i*2
        if i > len(array)-1:
            i = len(array) - 1
            break
    return binary_search(array, i//2, i, target)

array = [1, 2, 3, 5, 6, 8, 9, 10, 12]
for i in range(len(array)):
    print(f"Searching {array[i]}:\t position found = {exponential_search(array, array[i])}")
print(f"Searching 7:\t position found = {exponential_search(array, 7)}")
```
